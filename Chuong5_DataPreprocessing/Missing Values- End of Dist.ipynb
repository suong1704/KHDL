Thay thế bằng giá trị đuôi của phân bố dữ liệu 

```python
import pandas as pd
df=pd.read_csv('titanic.csv',usecols=['Age','Fare','Survived'])
df.head()
```

```python
df['Age'].isnull().sum()
```

```python
df.Age.hist(bins=50)
```

```python
# gia tri o duoi cua phan bo (bien Age theo phan bo chuan)
extreme = df.Age.mean() + 3*df.Age.std()
extreme
```

```python
import seaborn as sns
sns.boxplot('Age',data=df)
```

```python
# định nghia hàm thay thế dữ liệu theo 2 cách: Median & End of Distribution
def impute_nan(df,variable,median,extreme):
    df[variable+"_end_dist"]=df[variable].fillna(extreme)
    df[variable].fillna(median,inplace=True)
```

```python
# gọi hàm thay thế dữ liệu
impute_nan(df,'Age',df.Age.median(),extreme)
```
